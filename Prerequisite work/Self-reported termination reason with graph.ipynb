## Analysis of self-reported termination reason

```python
import pandas as pd
import numpy as np
df = pd.read_excel("ABS Tech Case 2026_Data.xlsx")
freq_table = df["TermReason"].value_counts().reset_index()
freq_table.columns = ["TermReason", "Count"]
freq_table["Percentage"] = freq_table["Count"] / freq_table["Count"].sum()

freq_table
```

```python
# Get df with only terminated employees
terminated = df[df["Termd"] == 1]

# Create a crosstab for counts and percentages
count_table = pd.crosstab(
    terminated["TermReason"],
    terminated["EmploymentStatus"]
)

percent_table = pd.crosstab(
    terminated["TermReason"],
    terminated["EmploymentStatus"],
    normalize="columns"
) * 100

# Sort by "Voluntarily Terminated" percentage
sorted_index = percent_table["Voluntarily Terminated"].sort_values(ascending=False).index
count_sorted = count_table.loc[sorted_index]
percent_sorted = percent_table.loc[sorted_index]

# Combine counts and percentages into a single table
final_table = count_sorted.astype(str) + " (" + percent_sorted.round(2).astype(str) + "%)"
final_table
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

palette = sns.color_palette("Set2", n_colors=13)
voluntary_percent = percent_sorted["Voluntarily Terminated"]
voluntary_percent = voluntary_percent[voluntary_percent > 0]
sns.barplot(
    x=voluntary_percent.values,
    y=voluntary_percent.index,
    hue=voluntary_percent.index,
    palette=palette,
    legend=False
)
plt.xlabel("Percentage (%)")
plt.ylabel("Termination Reason")
plt.show()
```
